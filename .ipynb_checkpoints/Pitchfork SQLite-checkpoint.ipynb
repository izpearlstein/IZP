```python
import sqlite3
import pandas as pd
```

```python
# Create the connection
conn = sqlite3.connect(r'database.sqlite')

# create the dataframe from a query
df = pd.read_sql_query("select reviews.reviewid, title, artist, url, score, best_new_music, pub_year, genre, content, label \
                        from reviews \
                        left join genres on reviews.reviewid = genres.reviewid \
                        left join content on reviews.reviewid = content.reviewid \
                        left join labels on reviews.reviewid = labels.reviewid \
                        where artist = 'massive attack' ", conn)
```
